```python
# 1. np.where 함수를 사용하여 평균이 60점 이상이면 pass, 아니면 fail 파생변수 만들기
# 파생변수명 result
np.where(exam_copy['평균']>=60, 'pass', 'fail')
exam_copy['result']=np.where(exam_copy['평균']>=60, 'pass', 'fail')
exam_copy
```

```python
# 2. apply 함수 사용하여 평균이 80 이상이면 A등급, 70 이상이면 B등급, 60이상이면 C등급
# 그 외 F 등급이 주어지는 파생변수 만들기
def grade(row):
    mean = row['평균']
    
    if mean >= 80:
        return "A"
    elif mean >= 70:
        return "B"
    elif mean >= 60:
        return "C"
    else:
        return "F"
    
   

# 파생변수명 grade
```

```python
exam_copy.apply(grade, axis = 1)
exam_copy['grade']=exam_copy.apply(grade, axis = 1)
exam_copy
```

```python
# 3. nclass 별 pass와 fail 학생수 확인하기
exam_copy.groupby(['nclass', 'result']).agg(nclass_people = ('result', 'count'))
```

```python
# 4. nclass 별 등급(A~F) 학생 수 확인
exam_copy.groupby(['nclass', 'grade']).agg(nclass_grade = ('grade', 'count'))
```

```python
# 번외 3번 예제 결과값은 fail이 먼저 나옴 pass / fail순으로 바꿔보는 작업
# 힌트 : sort_index
exam_copy.groupby(['nclass', 'result']).agg(nclass_people = ('result', 'count')).sort_index(ascending=False)
```

## Not a Number : 연산과정에서 잘못된 값을 입력 받았다

# 결측치
- 누락된 값, 비어있는 값
- 수집 과정에서 오류가 발생하거나 포함되어 있자 않는 값 가지고 올 때 
- 데이터의 오류 -> 분석 결과의 왜곡이 발생 
- 타이타닉 데이터 carbin 비어있는 데이터 

- 결측치 해결하는 방법 2가지
    - 1. 채워준다 -> 결측치의 개수가 적을 때 -> 평균값이나 중간값 채움
    - 2. 버린다 -> 결측 개수가 너무 많을 때 -> 컬럼 자체를 버림
                -> 결측치가 적을 때 
                
# 이상치
- 정상범위에서 크게 벗어난 값
- 논리적으로 존재할 수 없는 값
- 타이타닉에서 나이데이터 100

- 이상치 해결하는 방법 2가지
    - 1. 바꿔준다
    - 2. 버린다
